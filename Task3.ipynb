```python
#2 Count
# Returns the no of time perticular value is repeated in the list
#2.1 List of numbers
lis1=[1,2,3,4,5,1,1,1,2,3,3,3,3]
lis1.count(1) # Value '1' is repeted 4 times in the list
```

```python
lis1.count(3) # Value '3' is repeted 5 times in the list
```

```python
lis1.count(4) # Value '4' is repeted 1 time in the list
```

```python
lis1.count(6) # Value '6' is repeted 0 times in the list
```

```python
#2.2 List of vehicle brand
brand=["Tata","Suzuki","Honda","Hero","Tata","MG","Hero","Toyota","Hero"]
brand.count("Hero") # Hero is repeated 3 Times in list
```

```python
brand.count("Tata") # Tata is repeated 2 Times in list
```

```python
brand.count("Kia") # Kia is repeated 0 Times in list
```

```python
#3 Index
#Returns the position at the first occurrence of the specified value
#3.1
brand1=["Tata","Suzuki","Honda","Hero","Tata","MG","Hero","Toyota","Hero"]
brand1.index("Hero") # Hero is present at multiple position but returns the position at the first occurrence i.e 3rd position
```

```python
brand1.index("Toyota") # Toyota is present at the 7th position
```

```python
#3.2
lis3=[1,2,3,4,5,1,1,1,2,3,3,3,3]
lis3.index(2) # 2 is present at multiple position but returns the position at the first occurrence i.e 1st position
```

```python
lis3.index(1) # 1 is present at multiple position but returns the position at the first occurrence i.e 0th position
```

```python
lis3.index(4) # 4 is present at multiple position but returns the position at the first occurrence i.e 3rd position
```
